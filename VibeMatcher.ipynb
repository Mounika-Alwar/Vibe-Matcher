# Generating Embeddings

```python
# loading dataset
with open("products.json", "r") as f:
    products = json.load(f)

# loading embedding model
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# generating embeddings
for item in products:
    item["embedding"] = model.encode(item["description"]).tolist()

print("Embeddings created using sentence-transformers.")
```

```python
# Converting embeddings to a float32 numpy array
vectors = np.array([item["embedding"] for item in products]).astype("float32")

# Creating FAISS index
index = faiss.IndexFlatL2(vectors.shape[1])
index.add(vectors)

# Saving FAISS index
faiss.write_index(index, "vibe_index.faiss")

# Saving products metadata as JSON
with open("products.json", "w") as f:
    json.dump(products, f, indent=2)

print("FAISS index and metadata saved.")
```

```python
def search(query, top_k=3):
    q_emb = model.encode(query).astype("float32").reshape(1, -1)
    distances, indices = index.search(q_emb, top_k)
    results = []
    for dist, idx in zip(distances[0], indices[0]):
        item = products[idx]
        results.append({
            "name": item["name"],
            "description": item["description"],
            "vibes": item["vibes"],
            "distance": float(dist)
        })
    return results
```

# Cosine Similarity

```python
def cosine_similarity(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def search(query, top_k=3, threshold=0.30):
    q_emb = model.encode(query).astype("float32")

    # computing cosine similarity for all products
    scored = []
    for item in products:
        sim = cosine_similarity(q_emb, item["embedding"])
        scored.append((sim, item))

    # sorting by similarity descending
    scored.sort(key=lambda x: x[0], reverse=True)

    top = scored[:top_k]

    # fallback logic if best match is too weak
    if top[0][0] < threshold:
        return {
            "message": "No strong match found. Here are some available styles:",
            "suggest_categories": list({p["category"] for p in products})
        }

    return [
        {
            "name": item["name"],
            "description": item["description"],
            "vibes": item["vibes"],
            "category": item["category"],
            "similarity": float(sim)
        }
        for sim, item in top
    ]
```

# Testing Queries

```python
import time

# test queries
queries = [
    ("cozy sweater", "A"),
    ("Bossy peach blazer", "B"),
    ("Apple juice", "C")
]

results_log = []

for text, label in queries:
    start = time.time()
    result = search(text, top_k=3, threshold=0.30)
    end = time.time()
    latency = end - start

    # fallback case
    if isinstance(result, dict):
        results_log.append({
            "query_label": label,
            "query_text": text,
            "result_type": "fallback",
            "top_item": None,
            "similarity": None,
            "latency_seconds": latency
        })
        continue

    # normal match
    best_match = result[0]
    results_log.append({
        "query_label": label,
        "query_text": text,
        "result_type": "match",
        "top_item": best_match["name"],
        "similarity": best_match["similarity"],
        "latency_seconds": latency
    })

df_eval = pd.DataFrame(results_log)


# classify similarity quality
def classify_score(sim):
    if sim is None:
        return "no_match"
    if sim >= 0.7:
        return "good"
    elif sim >= 0.4:
        return "avg"
    else:
        return "bad"

df_eval["quality"] = df_eval["similarity"].apply(classify_score)

df_eval
```
